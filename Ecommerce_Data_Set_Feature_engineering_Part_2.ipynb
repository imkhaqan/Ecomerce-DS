```python
#before removing outliers
df.plot(kind="box",subplots=True,sharey=False,figsize=(22,10))
```

```python
df_num = df.select_dtypes(exclude=['object'])
def outlier_removal(data,col_name):
    Q1=data[col_name].quantile(0.25)
    Q3=data[col_name].quantile(0.75)
    IQR=Q3-Q1
    upper_limit=Q3+IQR*2
    lower_limit=Q1-IQR*2
    df_without_outlier=df.loc[(data[col_name]<=upper_limit)&(data[col_name]>=lower_limit)]
    return df_without_outlier

for col in df_num.columns:
    df_without_outlier =df.copy()
    #if col not in ['item_id','Year','Month','day','month','year']:
    df_without_outlier = outlier_removal(df_without_outlier, col)
df_without_outlier.shape
```

```python
df_without_outlier.plot(kind="box",subplots=True,sharey=False,figsize=(22,10))
```
