```python
# Feature engineering part here 1) Handling Null Values
df.isnull().sum()
```

```python
df.isnull().sum()
```

```python
#filling null values of status with pending
df['status'] = df['status'].fillna('pending')
```

```python
df.isnull().sum()
```

```python
# filling null values for Customer ID with Guest
df['Customer ID'] = df['Customer ID'].fillna('Guest')
```

```python
df['order_status'] = df['order_status'].fillna('pending')
```

```python
df['sku'] = df['sku'].fillna('unknown_SKU')
```

```python
df['sales_commission_code'] = df['sales_commission_code'].fillna('No-Code')
```

```python
df.isnull().sum()
```

```python
# Delete all rows ehere category name is null
df = df.dropna(subset=['category_name_1'])
```

```python
df['Customer Since'].fillna(df['created_at'], inplace=True)
```
